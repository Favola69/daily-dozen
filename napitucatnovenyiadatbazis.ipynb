```python
import pandas as pd

df = pd.read_csv('/content/napi_tucat_novenyi_adatbazis_bovitett.csv')

df.head()
```

```python
### 1. Adatok méretének vizsgálata
# Megmutatja az adatkeret sorainak és oszlopainak számát.
print("Az adatkeret mérete (sorok, oszlopok):", df.shape)
# Részletes információk az adatkeretről, beleértve az oszlopok adattípusait és a nem null értékek számát.
print("\nAdatkeret információk:")
df.info()

### 2. Oszlopok listázása
# Kilistázza az adatkeret összes oszlopneveit.
print("\nOszlopnevek:")
print(df.columns.tolist())

### 3. Hiányzó adatok elemzése
# Megszámolja a hiányzó (NaN) értékeket minden oszlopban.
print("\nHiányzó értékek oszloponként:")
print(df.isnull().sum())
# Megmutatja a hiányzó értékek százalékos arányát minden oszlopban.
print("\nHiányzó értékek százalékos aránya oszloponként (%):")
print((df.isnull().sum() / len(df) * 100).sort_values(ascending=False))
```

```python
import pandas as pd
df = pd.read_csv('/content/napi_tucat_novenyi_adatbazis_bovitett.csv')

print("Élelmiszerkategóriák nevei:")
print(df['Kategória'].unique())
```

```python
print("Kategóriák gyakorisága:")

print(
df["Kategória"]
.value_counts()
)
```

```python
print("Szezonok:")

print(
df["Szezon"]
.value_counts()
)
```

```python
print("Átlagos fehérje és rost kategóriánként:")

eredmeny = df.groupby("Kategória")[
["Fehérje_g_100g","Rost_g_100g"]
].mean()

print(eredmeny)
```

```python
import matplotlib.pyplot as plt

eredmeny.plot(
kind="bar",
figsize=(10,5)
)

plt.title("Fehérje és rost kategóriánként")
plt.xlabel("Kategória")
plt.ylabel("g / 100 g")

plt.xticks(rotation=45)

plt.show()
```

```python
import firebase_admin
from firebase_admin import credentials, firestore
import pandas as pd

# Firebase kapcsolat
cred = credentials.Certificate("/content/plantplateai-firebase-adminsdk.json")


db = firestore.client()

# CSV beolvasása
df = pd.read_csv('/content/napi_tucat_novenyi_adatbazis_bovitett.csv')

# Feltöltés Firestore-ba
for index, row in df.iterrows():
    data = row.to_dict()
    db.collection('foods').add(data)

print("Sikeres feltöltés!")
```
